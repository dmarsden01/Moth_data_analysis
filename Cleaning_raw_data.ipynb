## Reading .csvs

```python
# These "raw" data spreadsheets need tidying.
# They can also be joined together to a master dataframe.
Data_2021 = pd.read_csv('./2021_raw.csv', header=None)
Data_2022 = pd.read_csv('./2022_raw.csv', header=None)
Data_2023 = pd.read_csv('./2023_raw.csv', header=None)

Full_df = pd.read_csv('./VC62_Moth_Recording_PM.csv')

raw_data = {2021:Data_2021, 2022:Data_2022, 2023:Data_2023}

# Data_2021.head()
```

```python
def Remove_empty_heads(df_raw):
    """Takes the dataframe created from a raw csv with empty first row. Returns a dataframe with the first row being the corrected headers."""
    df_edited = df_raw[1:] #Empty first row removed
    df_edited = df_edited.reset_index(drop=True)
    
    return df_edited

def Remove_title_row_from_data(df):
    """Takes a df with index 0 being row of column titles. Also resets indices."""
    df.columns = df.iloc[0]
    df_edited = df.drop(index=(0))
    df_edited=df_edited.reset_index(drop=True) #Have indices start from 0 again.
    return df_edited

# def Simple_merge_csvs(df_dict):
#     """Given a dict of dataframes, will merge into one for all years."""
    
#     for df in df_dict:
```

```python
Cleaned_dfs = {}
for df in raw_data:
    Cleaned_dfs[df]=Remove_empty_heads(raw_data[df])
    Cleaned_dfs[df]=Remove_title_row_from_data(Cleaned_dfs[df])
    
```

## Merging

```python
for df in Cleaned_dfs:
    unique_species = set(Cleaned_dfs[df]["species"])
    print(f"{df} Species: {len(unique_species)}")
```

```python
df_merge = pd.merge(Cleaned_dfs[2021], Cleaned_dfs[2022], how="outer", on=["family", "species"])
df_merge_all = pd.merge(df_merge, Cleaned_dfs[2023], how="outer", on=["family", "species"])

df_merge_all.totals_x = pd.to_numeric(df_merge_all.totals_x, errors='coerce').fillna(0).astype(np.int64)
df_merge_all.totals_y = pd.to_numeric(df_merge_all.totals_y, errors='coerce').fillna(0).astype(np.int64)
df_merge_all.totals = pd.to_numeric(df_merge_all.totals, errors='coerce').fillna(0).astype(np.int64)

df_merge_all["Total"] = df_merge_all[["totals_x", 'totals_y', 'totals']].sum(axis=1)
```

```python
print(f"Length merge: {len(df_merge)}")
print(f"Columns merge: {len(df_merge.keys())}")

print(f"Length full merge: {len(df_merge_all)}")
print(f"Columns full merge: {len(df_merge_all.keys())}")
```
